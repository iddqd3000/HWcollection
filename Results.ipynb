```python
Для MyLinkendList, LinkedList, ArrayList, LinkedHashSet, LinkedHashMap были выполнены замеры времени операций вставки и удаления.
Измерения времени производилось 7 раз после чего высчитанно средне среднее арифметическое.

MyLinkedList add time: 3, 1, 1, 1, 0, 0, 2 // среднее - 1.14
LinkedList add time: 2, 2, 4, 1, 2, 4, 1 //среднее -  2.28
ArrayList add time: 1, 1, 0, 1, 0, 4, 9 //среднее -  2.28
LinkedHashSet add time: 3, 51, 8, 2, 2, 1, 2 //среднее -  9.71
LinkedHashMap add time: 3, 4, 1, 0, 6, 135, 1 //среднее -  21.42

MyLinkedList remove time:  102, 61, 30, 541, 40, 310, 148 //среднее -  176
LinkedList remove time:  134, 72, 29, 197, 519, 92, 139 //среднее -  168.85
ArrayList remove time:  10, 13, 8, 63, 13, 17, 7 //среднее -  18.71
LinkedHashSet remove time:  3, 0, 1 ,0, 0, 1, 0 //среднее -  0.71
LinkedHashMap remove time:  1, 0, 1, 0, 0, 1, 0 //среднее -  0.42
```

```python
import matplotlib.pyplot as plt
import numpy as np 
cat_par = [f"Коллекция {i}" for i in range(5)]
Add_time = [1.14, 2.28, 2.28, 9.71, 21.42] 
Remove_time = [176,168.85, 18.71, 0.71, 0.42]
width = 0.3
x = np.arange(len(cat_par))
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Add_time, width, label='Add time')
rects2 = ax.bar(x + width/2, Remove_time, width, label='Remove time')
ax.set_title('Add/Time diagram')
ax.set_xticks(x)
ax.set_xticklabels(cat_par)
ax.legend()
plt.show()
```

```python
Колекция0- MyLinkendList 
Колекция1- LinkedList
Колекция2- ArrayList
Колекция3- LinkedHashSet
Колекция4- LinkedHashMap

Время в милисекундах.
```
